4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)

```python
advogados = df[df['Profissão'].isin(['Advogado(a)'])]
repre_vendas = df[df['Profissão'].isin(['Representante de Vendas'])]
todos = df['Duração do sono'].mean()

sono_advogados = advogados['Duração do sono'].mean()
sono_repre_vendas =  repre_vendas['Duração do sono'].mean()

print(f'A média de sono entre advogaos é: {sono_advogados:.2f}\nE entre Representantes de Venda é: {sono_repre_vendas}\nA média de sono entre todos os profissionais é {todos:.2f}')
```

**apenas Representantes de Vendas dormem a baixo da média**

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)

```python
medicos = df[df['Profissão'].isin(['Médico(a)'])]
enfermeiros = df[df['Profissão'].isin(['Enfermeiro(a)'])]

sono_medicos = medicos['Duração do sono'].mean()
sono_enfermeiros =  enfermeiros['Duração do sono'].mean()

menos_horas_sono = np.where(sono_enfermeiros < sono_medicos, 'Enfermeiro(a)', 'Médico(a)')
print(f"o profissional com menos horas de sono é o {menos_horas_sono}")
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
df_pressao = df.copy()
df_pressao = df_pressao.drop(columns = ['Identificador', 'Nível de atividade física','Duração do sono', 'Distúrbio do sono', 'Frequência cardíaca', 'Passos diários', 'Categoria IMC', 'Qualidade do sono' ])
df_pressao[['Pressão Sistólica', 'Pressão Diastólica']] = df_pressao['Pressão Sanguinea'].str.split('/', expand=True)
df_pressao.head()
```

```python
pressao_sistolica = df_pressao.groupby('Gênero')['Pressão Sistólica'].mean()
pressao_diastolica = df_pressao.groupby('Gênero')['Pressão Diastólica'].mean()

maior_pressao_sistolica = np.where(pressao_sistolica['Mulher'] > pressao_sistolica['Homem'], 'Mulher', 'Homem')
maior_pressao_diastolica = np.where(pressao_diastolica['Mulher'] > pressao_diastolica['Homem'], 'Mulher', 'Homem')

print(f'A população de {maior_pressao_sistolica} tem a maior pressao_diastolica sendo assim {maior_pressao_diastolica}s tem a maior pressão diastolica por consequência')
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
frec_acima = df[df['Frequência cardíaca'] > 70]
frec_menor_igual = df[df['Frequência cardíaca'] <= 70]

passos_frec_acima = frec_acima['Passos diários'].mean()
passos_frec_menor_igual = frec_menor_igual['Passos diários'].mean()

mais_passos = np.where(passos_frec_acima > passos_frec_menor_igual, ' acima de 70', 'menor ou igual a 70')

print(f'pessoas com frequências cardíacas {mais_passos} dão mais passos')
```
